Imports

```python
from neo4j import GraphDatabase
```

Connect

```python
uri= "bolt://localhost:7687"
username= "neo4j"
password= "neo4j"


graphDB_Driver = GraphDatabase.driver(uri, auth=(username, password))
```

Partie 1 : Create all function needed

```python
def get_all_nodes():
    with graphDB_Driver.session() as session:
        result = session.run("MATCH (n) RETURN n")
        for record in result:
            print(record)
    

# get_all_nodes()
```

```python
def get_node_by_id(id):
    with graphDB_Driver.session() as session:
        result = session.run("MATCH (n) WHERE id(n) = $id RETURN n", id=id)
        for record in result:
            print(record)
            
get_node_by_id(0)
```

```python
def get_node_by_label(label):
    with graphDB_Driver.session() as session:
        result = session.run("MATCH (n) WHERE $label IN labels(n) RETURN n", label=label)
        for record in result:
            print(record)

get_node_by_label("Person")
```

```python
def get_all_relationships():
    with graphDB_Driver.session() as session:
        result = session.run("MATCH (n)-[r]->(m) RETURN r")
        for record in result:
            print(record)
            
get_all_relationships()
```

```python
def get_relationship_by_id(id):
    with graphDB_Driver.session() as session:
        result = session.run("MATCH (n)-[r]->(m) WHERE id(r) = $id RETURN r", id=id)
        for record in result:
            print(record)   
            
get_relationship_by_id(0)
```

```python
def get_relationship_by_type(type):
    with graphDB_Driver.session() as session:
        result = session.run("MATCH (n)-[r]->(m) WHERE type(r) = $type RETURN r", type=type)
        for record in result:
            print(record)
```

```python
def create_node(label,properties):
    with graphDB_Driver.session() as session:
        session.run(f"CREATE (:{label} {properties})")
            
create_node("Person", '{born:1969, name: "Maurice Pintusson"}')
#We saw that the node is created 
get_node_by_label("Person")
```

```python
def create_relationship(start_node_id, end_node_id, type, properties):
    with graphDB_Driver.session() as session:
        session.run(f"MATCH (n1),(n2) WHERE id(n1) = $start_node_id AND id(n2) = $end_node_id CREATE (n1)-[:{type} {properties}]->(n2)", start_node_id=start_node_id, end_node_id=end_node_id)

create_relationship(177, 178, "KNOWS", "{since: 1999}")
#<Record r=<Relationship element_id='283' nodes=(<Node element_id='177' labels=frozenset() properties={}>, <Node element_id='178' labels=frozenset() properties={}>) type='KNOWS' properties={'since': 1999}>>
get_all_relationships()
```
